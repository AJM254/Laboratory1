Problem 5:

```python
def find_month(x):
    if " 01 " in x:
        return "Jan"
    elif " 02 " in x:
        return "Feb"
    elif " 03 " in x:
        return "March"    
    elif " 04 " in x:
        return "April"    
    elif " 05 " in x:
        return "May"    
    elif " 06 " in x:
        return "June"    
    elif " 07 " in x:
        return "July"    
    elif " 08 " in x:
        return "August"    
    elif " 09 " in x:
        return "Sep"    
    elif " 10 " in x:
        return "Oct"    
    elif " 11 " in x:
        return "Nov"    
    else:
        return "Dec"
```

```python
df["Month"]=df["Date/Time"].apply(find_month)
```

```python
df.groupby('Month').mean()
```

The months with the highest average wind speed are August, March and November. 

The months with the highest power output are August, March, and November. 

I expected this because the highest wind speed should have the highest power output in order to generate it.

Problem 1a: 

```python
df["Loss"] = df["LV ActivePower (kW)"] - df["Theoretical_Power_Curve (KWh)"]
print(df)
```

```python
df["Loss Percentage"] = (df["LV ActivePower (kW)"]/  df["Theoretical_Power_Curve (KWh)"])*100
print(df)
```

These Loss column represents the differenece in power between actual power (generated) and the theoretical power (what should be generated) when the wind turbine is used. The loss percentage is the per power unit generated in loss between the actual power and the theoretical. 

Problem 1b:

```python
x = df["Wind Speed (m/s)"]
y = df["Loss"]
plt.scatter(x,y)
plt.title("Wind Speed versus Loss")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Loss")
```

```python
x = df["Wind Speed (m/s)"]
y = df["Loss Percentage"]
plt.scatter(x,y)
plt.title("Wind Speed versus Loss Percentage")
plt.xlabel("Wind Speed (m/s)")
plt.ylabel("Loss Percentage")
```

The wind speed affects the loss because at lower wind speeds there is a signficant power loss compared to the higher wind speeds. The loss is greatest between wind speeds 0-17 m/s.
